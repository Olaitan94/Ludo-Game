This is a simple dice rolling game.

How the game works:
 -- Multi player game (min 2 and max 4 players allowed
 -- Roll a 6 to start.
 -- Keep rolling to continue
 -- First person to reach 50 wins.

Algorithm
1. Welcome message (line 18)
2. Select number of players( lines 19 - 24)
3. Players registration (lines 25 - 33)
4. Start the game 
5. game will be played in a while loop which will only be broken when one player reaches 50.
6. print results 
7. plot results

```python
import random
import matplotlib.pyplot as plt

%matplotlib inline


def dice(num):
    if num == 1:
        print('[     ]\n\
[  0  ]\n\
[     ]')
    
    elif num ==2:
        print('[     ]\n\
[ 0 0 ]\n\
[     ]')
    
    elif num ==3:
        print('[  0  ]\n\
[ 0 0 ]\n\
[     ]')
    
    elif num ==4:
        print('[  0  ]\n\
[ 0 0 ]\n\
[  0  ]')
    
    elif num==5:
        print('[  0  ]\n\
[ 0 0 ]\n\
[ 0 0 ]')

    elif num==6:
        print('[ 0 0 ]\n\
[ 0 0 ]\n\
[ 0 0 ]')
    

class Players(object):
    
    def __init__(self,name):
        
        self.name = name
        self.moves = []
        
    def roll(self):
        x = random.randint(1,6)
        return x 

print('Lets play Ludo!!')
no = input('Enter number of players (e.g 2 or 3):>> ')
while not no.isdigit() or int(no) < 2 or int(no) > 4:    #This will make sure only a digit is inputed.
    print('You have to enter an integer e.g 2. minimum of 2 and maximum of 4 players are allowed.')
    no = input('Enter number of players (e.g 2 or 3):>> ')
    

no = int(no)
player_list =[]
for n in range(no):
    player = input(f'Player{n+1} enter your name:>> ')
    player_list.append(player)

new = []
for n in player_list:
    n = Players(n)
    new.append(n)
    
z = 0
h = True
while h is True:
    
    for player in new:
        print(f'{player.name}, Get Ready!')
        a = player.roll()
        dice(a)
        if len(player.moves) == 0:
            while a != 6:
                print('Sorry, You nneed to play a 6 to start.Roll again.')
                a = player.roll()
                dice(a)
            print('Great! You rolled a 6!! Now roll again to continue.')
            a = player.roll()
            dice(a)
            player.moves.append(a)
        else:
            j = player.moves[-1]
            z = j + a
            player.moves.append(z)
            
        if z >= 50:
            h = False
            break
        else:
            continue
        
for player in new:
    print(f'{player.name} >> {player.moves}')
    if player.moves[-1] >= 50:
        print(f'{player.name} has won!')
    
    
for i in range(len(new)):
    plt.subplot(1,no,i+1), plt.plot(new[i].moves,'-bo')
    plt.title(new[i].name)
    plt.show
    
    
    
        
        

    



```
